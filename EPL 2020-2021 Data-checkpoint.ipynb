```python
# Display basic statistics 
print("Descriptive Statistics:") 
print(df.describe()) 
```

```python
# Player Performance Analysis 
top_scorers = df[['Name', 'Club', 'Goals', 'Assists', 
'Perc_Passes_Completed', 'xG']].sort_values(by='Goals', 
ascending=False) 
print("\nTop 10 Players by Goals:") 
print(top_scorers.head(10)) 
```

```python
# Team Contributions 
team_goals = df.groupby('Club')['Goals'].sum().sort_values(ascending=False) 
plt.figure(figsize=(12, 6)) 
sns.barplot(x=team_goals.index, y=team_goals.values, palette='viridis') 
plt.xticks(rotation=90) 
plt.title('Total Goals per Team') 
plt.xlabel('Team') 
plt.ylabel('Goals') 
plt.show() 
```

```python
# Positional Performance 
position_analysis = df.groupby('Position')[['Goals', 'Assists', 
'Perc_Passes_Completed']].mean() 
print("\nPerformance by Position:") 
print(position_analysis) 
```

```python
# Disciplinary Analysis 
yellow_red_cards = df.groupby('Position')[['Yellow_Cards', 'Red_Cards']].mean() 
print("\nDisciplinary Records by Position:") 
print(yellow_red_cards)
```

```python
# Passing Efficiency 
top_passers = df[['Name', 'Club', 
'Perc_Passes_Completed']].sort_values(by='Perc_Passes_Completed', 
ascending=False) 
print("\nTop 10 Players by Passing Accuracy:") 
print(top_passers.head(10))
```

```python
# xG vs Actual Goals 
plt.figure(figsize=(8, 6)) 
sns.scatterplot(x=df['xG'], y=df['Goals']) 
plt.xlabel('Expected Goals (xG)') 
plt.ylabel('Actual Goals') 
plt.title('xG vs Actual Goals') 
plt.show() 
```

```python
# Age Impact on Performance 
plt.figure(figsize=(8, 6)) 
sns.scatterplot(x=df['Age'], y=df['Goals']) 
plt.xlabel('Age') 
plt.ylabel('Goals') 
plt.title('Age vs Goals') 
plt.show()
```

```python
# Goalkeeper Performance 
goalkeepers = df[df['Position'] == 'GK'][['Name', 'Club', 
'Perc_Passes_Completed']] 
print("\nGoalkeeper Passing Accuracy:") 
print(goalkeepers.sort_values(by='Perc_Passes_Completed', ascending=False))
```

```python
# Correlation Analysis 
correlation_matrix = df[['Goals', 'Assists', 'Perc_Passes_Completed', 
'xG']].corr() 
plt.figure(figsize=(8, 6)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5) 
plt.title('Correlation Matrix') 
plt.show() 
```

```python
# Clustering Players based on Goals & Assists
X_cluster = df[['Goals', 'Assists']]
kmeans = KMeans(n_clusters=3)
df['Cluster'] = kmeans.fit_predict(X_cluster)
sns.scatterplot(x=df['Goals'], y=df['Assists'], hue=df['Cluster'],
palette='viridis')
plt.xlabel('Goals')
plt.ylabel('Assists')
plt.title('Player Clusters based on Goals & Assists')
plt.show()
```
